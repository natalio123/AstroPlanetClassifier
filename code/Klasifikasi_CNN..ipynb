```python
import os
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
TRAIN_DIR = 'C:/DataAnalysis/Projek/Projek Machine Learning/Klasifikasi_Benda_Langit/dataset/train'
```

```python
valid_extensions = ('.jpg', '.jpeg', '.png')
planets = ['Jupiter', 'Mars', 'Merkurius', 'Neptunus', 'Saturnus', 'Uranus', 'Venus']

fig, axes = plt.subplots(1, len(planets), figsize=(15, 5))
if len(planets) == 1: 
    axes = [axes]

for ii, planet in enumerate(planets):
    dir = f'C:/DataAnalysis/Projek/Projek Machine Learning/Klasifikasi_Benda_Langit/dataset/train/{planet}'
    if not os.path.exists(dir):
        print(f"Folder '{dir}' tidak ditemukan.")
        continue
    files = [f for f in os.listdir(dir) if f.lower().endswith(valid_extensions)]
    if not files:
        print(f"Folder '{dir}' tidak memiliki file gambar yang valid.")
        continue
    try:
        img = tf.keras.preprocessing.image.load_img(dir + '/' + files[0])
        axes[ii].imshow(img)
        axes[ii].set_title(f'Ilustrasi Planet {planet}')
        axes[ii].axis('off')
    except Exception as e:
        print(f"Gagal memuat gambar dari '{dir}': {e}")
plt.show()
```
